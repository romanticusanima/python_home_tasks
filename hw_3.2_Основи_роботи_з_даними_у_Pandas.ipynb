```python
# 1 гіпотеза: середня сума інвойсу клієнтів з Mandalay вища
yangon_invoice_count = df.query('City == "Yangon"').shape[0]
mandalay_invoice_count = df.query('City == "Mandalay"').shape[0]
yangon_invoice_price_mean = df.query('City == "Yangon"')['Unit price'].mean()
mandalay_invoice_price_mean = df.query('City == "Mandalay"')['Unit price'].mean()
print(f"кількість інвойсів у Yangon {yangon_invoice_count} з середньою сумою інвойсу {yangon_invoice_price_mean:.2f}")
print(f"кількість інвойсів у Mandalay {mandalay_invoice_count} з середньою сумою інвойсу {mandalay_invoice_price_mean:.2f}")
```

9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.
