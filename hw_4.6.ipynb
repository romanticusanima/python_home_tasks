```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("yulu_rental.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='windspeed', y='count', alpha=0.5, c='#ab96cc')
plt.title('Impact of Windspeed on the Count of Rented Bikes')
plt.xlabel('Windspeed')
plt.ylabel('Total Count')
plt.show()
```

```python
season_rentals = df.groupby('season')['count'].sum()
season_rentals
```

```python
season_rentals.plot.pie(labels=['Spring', 'Summer', 'Autumn', 'Winter'], autopct='%1.1f%%', colors=['#32b31f', '#eff122', '#d37b1f', '#387be3']);
```

```python
plt.figure(figsize=(10, 6))
winter_data = df.query('season == 4').groupby('temp')['registered'].sum()
winter_data.plot.bar();
```
